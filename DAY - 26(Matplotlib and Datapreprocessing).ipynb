```python
x = ['x','y','z','a','b']
y = [12,37,89,45,67]
plt.bar(x,y,color=['g','black','b','cyan'],width=0.5)
plt.xlabel("X")
plt.ylabel("Y")

plt.title("bar plot")

plt.show()
```

```python
# horizontal plot
x = ['x','y','z','a','b']
y = [12,37,89,45,67]
plt.barh(x,y,color=['g','black','b','cyan'])
plt.xlabel("X")
plt.ylabel("Y")
plt.text(12,'x','pandas')
plt.text(37,'y','numpy')
plt.text(89,'z','math')
plt.text(45,'a','scikitlearn')
plt.text(67,'b','python')
plt.title("bar plot")
plt.show()
```

```python
marks = [23,45,60,56,78]
plt.pie(marks,labels=['Physics','Maths1',"Maths2",'Chemistry','Python'])

plt.show()
```

```python
marks = [23,45,60,56,78]
plt.pie(marks,labels=['Physics','Maths1',"Maths2",'Chemistry','Python'],
       explode=[0,0.25,0.1,0.1,0.01])

plt.show()
```

```python
marks = [23,45,60,56,78]
plt.pie(marks,labels=['Physics','Maths1',"Maths2",'Chemistry','Python'],
       explode=[0,0.25,0.1,0.1,0.01],labeldistance=0.5)

plt.show()
```

```python
marks = [23,45,60,56,78]
plt.pie(marks,labels=['Physics','Maths1',"Maths2",'Chemistry','Python'],
       explode=[0,0.25,0.1,0.1,0.01])
plt.show()
```

```python
marks = [23,45,60,56,78]
plt.pie(marks,labels=marks,radius=1.0,labeldistance=0.5)

plt.show()
```

```python
marks = [23,45,60,56,78]
plt.pie(marks,labels=marks,radius=1.5,labeldistance=0.5,
       colors=['r','g','c','y','m'])

plt.show()
```

#### Standard scalar

- mean=0 std=1
- z=x-mue/sigma
- z=x-mean/std

$$ z=(x-\mu) / \sigma $$

```python
from sklearn.preprocessing import StandardScaler
```

```python
s = StandardScaler()
std_data = s.fit_transform(df) # convert df data into standard scalar distribution data like mean=0 std=1
std_data
```
