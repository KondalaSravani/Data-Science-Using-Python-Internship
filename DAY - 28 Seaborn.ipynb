````python
import matplotlib.pyplot as plt
````

````python
plt.subplot(1,2,1)# rows,col,index/position
plt.plot([1,2,3],[2,3,4])
plt.subplot(1,2,2)
plt.scatter([1,2,3,4,5],[5,6,7,8,9])
````

````python
import numpy as np
plt.subplot(2,1,1)
plt.bar(['x','y','z'],height=5)
plt.subplot(2,1,2)
plt.hist(np.random.randint(1,10,2),bins=5)
````

````python
x = np.random.randint(1,100,10)
y = np.random.randint(1,100,10)
plt.subplots(2,2)
plt.subplot(2,2,1)
plt.plot(x,y,'r*',label='line plot')
````

````python
x = np.array([1,2,3,4,5])
y = np.array([1,4,7,10,15])
plt.subplots(2,2)
plt.subplot(2,2,1)
plt.plot(x,x,'r*',label="line plot")
plt.subplot(2,2,2)
plt.scatter(x,y,label="scatter plot")
plt.subplot(2,2,3)
plt.bar(x,y,label="bar plot")
plt.subplot(2,2,4)
plt.barh(x,y,label="bar horizontal")
plt.legend()
````

````python
plt.subplot(1,2,1)
plt.pie([20,30,40])
plt.subplot(1,2,2)
plt.barh([20,30,40],width=5,color=['r','m','k'])
````

````python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
plt.plot(np.random.randn(50).cumsum(),'k',label='line plot')
ax2 = fig.add_subplot(2,2,2)
plt.hist(np.random.randn(100),bins=20,color='y',orientation='horizontal',label="bar plot")
ax3 = fig.add_subplot(2,2,3)
plt.scatter(np.arange(30),np.random.randint(1,50,30),label="scatter")
plt.legend()
plt.show()
````
