```python
import pandas as pd 

url = "https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Stock/MSFT_data.csv"
```

```python
df=pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(25,10))

df1 = df.head(60)
plt.plot(df1['date'],df1['open'])
plt.xlabel('Date')
plt.ylabel("Opening Price of the market")
plt.title("Date vs Opening Price")
plt.show()
```

```python
plt.figure(figsize=(25,10))

df1 = df.head(60)
plt.plot(df1['date'],df1['open'])
plt.xlabel('Date')
plt.ylabel("Opening Price of the market")
plt.title("Date vs Opening Price")

plt.xticks(df1['date'],rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(25,10)) #(width,height)

df1 = df.head(60)
df2 = df.tail(60)
plt.plot(df1['open'])
plt.plot(df2['open'])
plt.xlabel('Date')
plt.ylabel("Opening Price of the market")
plt.title("Date vs Opening Price")

#plt.xticks(df1['date'],rotation=90)
plt.show()
```

```python
x = np.arange(-55,55)
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.title("Y=F(X)")
plt.show()
```

```python
help(plt.plot)
```

```python
x = np.arange(-5,10)
plt.plot(x,x**2,c='black',marker='>')
plt.plot(x,x**3,c='#00ff00',marker='*')
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.title("Y=F(X)")
plt.show()
```

```python
x = np.arange(-5,10)
plt.plot(x,x**2,c='black',marker='>',linestyle='-.')
plt.plot(x,x**3,c='#00ff00',marker='*',linestyle='-.')
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.title("Y=F(X)")
plt.legend(['Y = X**2','Y = X**3'],loc='upper center')
plt.text(1,1, "(1,1)")
plt.text(5, 25, "(5, 25)")
plt.grid()
plt.show()
```

```python
x = np.arange(-5,10)
plt.plot(x,x**2,c='black',marker='>',linestyle='-.')
plt.plot(x,x**3,c='#00ff00',marker='*',linestyle='-.')
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.title("Y=F(X)")
plt.legend(['Y = X**2','Y = X**3'],loc='upper center')
plt.text(1,1, "(1,1)")
plt.text(5, 25, "(5, 25)")
plt.grid()
plt.axis("off")
plt.show()
```

### Scatter Plot

- identifying the relationship between two features

- +ve
- -ve
- neutral

```python
x = [1,2,3,4,5]
y = np.array([2,4,6,8,10])
y2=-y
y3=[5,5,5,5,5]
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,marker='>',c=['b','r','g','black','cyan'],alpha=0.9,s=y**2)
plt.show()
```

```python
plt.scatter(x,y,marker='>',c=['b','r','g','black','cyan'],alpha=0.9,s=y**5)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x,y,marker='>',c=['b','r','g','black','cyan'],alpha=0.9,s=y**3,edgecolors='black')
plt.grid()
plt.show()
```
