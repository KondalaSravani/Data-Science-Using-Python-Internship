### Box Plot

- IQR range
- min,median,max
- It is also called as whisker plot
- To show distributions w.r.t categories
- Syntax : seaborn.boxplot

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
df = sns.load_dataset("titanic")
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(x='age',y='survived',data=df)
```

```python
sns.boxplot(x='embarked',y='survived',data=df,hue='sex')
```

```python
df1 = sns.load_dataset("tips")
df1.head()
```

```python
sns.boxplot(data=df1)
```

```python
import seaborn as sns
#sns.set_theme("notebook")
sns.boxplot(x='tip',y='sex',data=df1,hue="smoker")
```

```python

sns.boxplot(x='tip',y='sex',data=df1,hue="smoker",palette="Set2",linewidth=5)
```

```python

sns.boxplot(x='tip',y='sex',data=df1,hue="day",palette="Set2",linewidth=5)
```

```python

sns.boxplot(data=df1,orient='h')
```

### Kde Plot

- Kernal Density Estimator
- sns.kdeplot()

```python
import seaborn as sns
```

```python
df1
```

```python
sns.kdeplot(data=df1)
```

```python
sns.kdeplot(x="tip",y="total_bill",data=df1,hue='sex')
```

```python
sns.kdeplot(x='tip',data=df1,bw_adjust=5,hue='day')
```

```python
sns.kdeplot(x='tip',data=df1,bw_adjust=5,hue='day',multiple='stack')
```

### violin Plot

- Box Plot + KDE Plot

```python
import seaborn as sns
```

```python
sns.violinplot(data=df1)
```

```python
sns.violinplot(x=df1['tip'],data=df1)
```

```python
sns.violinplot(x=df1['day'],y=df1['size'],data=df1)
```

```python
sns.violinplot(x=df1['day'],y=df1['size'],data=df1,hue='sex')
```

```python
sns.violinplot(x=df1['day'],y=df1['size'],data=df1,palette='Oranges')
```

```python
sns.violinplot(x=df1['day'],y=df1['size'],data=df1,palette='Oranges',split=True,hue='sex')
```

```python
sns.violinplot(x=df1['day'],y=df1['size'],data=df1,hue='sex',palette='Oranges',split=True,inner='quartile')
```

### Bar Plot

```python
sns.barplot(data=df1)
```

```python
sns.barplot(x='day',y='tip',data=df1,hue='smoker')
```

```python
import numpy as np
sns.barplot(x='day',y='tip',data=df1,estimator=np.mean)
```

```python
sns.barplot(x='day',y='tip',data=df1,estimator=np.std)# estimator takes only numeric data
```

```python
df1['tip'].std()
```

### Point Plot

```python
sns.pointplot(data=df1)
```

```python
sns.pointplot(x='day',y='tip',data=df1,estimator=np.mean)
```

```python
sns.pointplot(x='sex',y='total_bill',data=df1,estimator=np.mean)
```

```python
sns.pointplot(x='sex',y='total_bill',data=df1,hue='day',estimator=np.mean)
```
