```python
d1 = pd.DataFrame({'clg':['abc','GVP','GVPW','XYZ'],
                  'temp':[34,56,78,54],
                   'humidity':[45,23,78,89]})
d1
```

```python
d2 = pd.DataFrame({'clg':['abc','GVP','GVPW','XYZ'],
                  'temp':[14,46,68,44],
                   'humidity':[25,43,88,79]})
d2
```

```python
d3 = pd.concat((d1,d2))
d3
```

```python
d3 = pd.concat((d1,d2),ignore_index=True)
d3
```

### Merging DataFrames

- we get only common values

```python
a = pd.DataFrame({'x':[1,2,3,4],'y':[3,41,55,66]})
b = pd.DataFrame({'x':[1,2,3,4],'y':[3,4,5,6]})
```

```python
a
```

```python
b
```

```python
c=pd.merge(a,b)
c
```

```python
m1 = pd.DataFrame({"city":['X','Y','Z'],
                  "temp":[11,22,33]})
m1
```

```python
m2 = pd.DataFrame({"city":['X','Y','Z'],
                  "hum":[11,22,33]})
m2
```

```python
m3 = pd.merge(m1,m2)
m3
```

```python
m3 = pd.merge(m1,m2,on='city',how='left',indicator=True)
m3
```
